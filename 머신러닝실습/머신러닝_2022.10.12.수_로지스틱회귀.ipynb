```python
from sklearn.datasets import load_breast_cancer # 내장 데이터를 불러올 수 있는 환경 설정
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
cancer = load_breast_cancer()
```

```python
x_train, x_test, y_train, y_test = train_test_split(cancer.data, cancer.target, 
                                                    stratify = cancer.target, random_state = 0)
```

```python
lr = LogisticRegression(max_iter=5000) # max_iter는 층분히 큰 수여야 한다.
```

```python
lr.fit(x_train, y_train)
```

```python
train_score = lr.score(x_train, y_train)
test_score = lr.score(x_test, y_test)
```

```python
print(train_score)
print(test_score)
```

```python
lr100 = LogisticRegression(C=100, max_iter=500) 
# 로지스틱리그레션에서의 C값은 알파와 역수의 관계에 있다. 
# C 값이 커지면 규제가 작아진다.
# 알파값이 커지면 규제 또한 커진다.
# 규제가 커지면 과대적합을 조절할 수 있다.
```

```python
lr100.fit(x_train, y_train)
```

```python
train_score100 = lr100.score(x_train, y_train)
test_score100 = lr100.score(x_test, y_test)
```

```python
print(train_score100)
print(test_score100)
```

```python
lr001 = LogisticRegression(C=0.01, max_iter=5000)
lr001.fit(x_train, y_train)
```

```python
train_score001 = lr001.score(x_train, y_train)
test_score001 = lr001.score(x_test, y_test)
```

```python
print(train_score001)
print(test_score001)
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.plot(lr100.coef_.T, '^', label='C=100')
plt.plot(lr.coef_.T, 'o', label='C=1')
plt.plot(lr001.coef_.T, 'v', label='C=0.01')
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.ylim(-5, 5)
plt.xlabel('특성')
plt.ylabel('계수크기')
plt.legend()
plt.show()
```

```python
for C, marker in zip([0.01, 1, 100], ['o', '^', 'v']):
    lr_l1 = LogisticRegression(solver='liblinear', C=C, penalty='l1', 
                               max_iter = 1000).fit(x_train, y_train)
    print('C={:.3f}인 l1 로지트식 회귀의 훈련 정확도: {:.2f}'.format(
        C, lr_l1.score(x_train, y_train)))
    print('C={:.3f}인 l1 로지트식 회귀의 테스트 정확도: {:.2f}'.format(
        C, lr_l1.score(x_test, y_test)))
    plt.plot(lr_l1.coef_.T, marker, label='C={:.3f}'.format(C))
    
plt.xticks(range(cancer.data.shape[1]), cancer.feature_names, rotation=90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.xlabel('특성')
plt.ylabel('계수크기')
plt.ylim(-5, 5)
plt.legend(loc=3)
plt.show()
```
