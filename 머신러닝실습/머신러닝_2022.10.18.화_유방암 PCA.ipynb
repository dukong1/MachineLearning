```python
import numpy as np
import pandas as pd 
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import mglearn
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
cancer = load_breast_cancer()
```

### PCA 적용해서 유방암 데이터셋 시각화하기

```python
fig, axes = plt.subplots(15, 2, figsize=(10, 20))

malignant = cancer.data[cancer.target==0] # 악성
benign = cancer.data[cancer.target==1] # 양성

ax = axes.ravel() # 차원축소

for i in range(30):
    _, bins = np.histogram(cancer.data[:,i], bins=50)
    ax[i].hist(malignant[:,i], bins=bins, color=mglearn.cm3(0), alpha=.5)
    # 악성
    ax[i].hist(benign[:,i], bins=bins, color=mglearn.cm3(2), alpha=.5)
    # 양성
    ax[i].set_title(cancer.feature_names[i]) 
    # 각각의 피쳐네임으로 그래프 이름 지정
    ax[i].set_yticks(())
    
ax[0].set_xlabel('특성크기')
ax[0].set_ylabel('빈도')
ax[0].legend(['악성', '양성'], loc='best')
fig.tight_layout()
```

### 양성, 악성 별 주성분 그래프 그려보기

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(cancer.data)
x_scale = scaler.transform(cancer.data)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2) # 성분 두 개를 뽑는다.
pca.fit(x_scale)
```

```python
# 처음 두 개의 주성분을 사용해서 데이터 변환
x_pca = pca.transform(x_scale)
```

```python
print('원본 데이터 형태: ', str(x_scale.shape))
print('축소된 데이터 형태: ', str(x_pca.shape)) # 주성분의 형태
```

```python
# 클래스를 색깔로 구분해서 처음 두 개 주성분을 그래프로 나타내기

plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(x_pca[:,0], x_pca[:,1], cancer.target)
plt.legend(['악성', '양성'], loc='best') # legend() -> 범례
plt.xlabel('첫 번째 주성분')
plt.ylabel('두 번째 주성분')
plt.show()
```

```python
print('주성분: ', pca.components_)
```
